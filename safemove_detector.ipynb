```python
results = model("test.jpg", conf=0.3)

animal_classes = [15, 16, 17, 18, 19]
# COCO IDs:
# 15 = cat
# 16 = dog
# 17 = horse
# 18 = sheep
# 19 = cow

animal_found = False

for box in results[0].boxes:
    cls = int(box.cls[0])
    if cls in animal_classes:
        animal_found = True

if animal_found:
    print("🟥 ANIMAL DETECTED")
else:
    print("🟩 NO ANIMAL")
```
